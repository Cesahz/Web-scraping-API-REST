# Ecosistema Open Library

## Endpoint 1A: El Buscador Principal (/search.json)

Propósito: Encontrar el ID único del autor a partir del título limpio del libro.

Método HTTP: GET

URL Base: https://openlibrary.org/search.json

Dato que devuelve con parametros: title, limit y fields

```
{
  "numFound": 21,
  "start": 0,
  "numFoundExact": true,
  "num_found": 21,
  "documentation_url": "https://openlibrary.org/dev/docs/api/search",
  "q": "",
  "offset": null,
  "docs": [
    {
      "author_key": [
        "OL7527945A"
      ],
      "author_name": [
        "Akiko Higashimura"
      ],
      "title": "Princess jellyfish volume 1"
    }
  ]
}
```
## Endpoint 1B: Detalles del Autor (/authors/{ID}.json)

Propósito: Extraer la fecha de nacimiento cruda.

Método HTTP: GET

URL Base: https://openlibrary.org/authors/OL3418464A.json

Nota: usa Path Variables. El ID se inyecta directamente en la URL.

Configuración en Thunder Client:

```
{
  "name": "Akiko Higashimura",
  "personal_name": "Akiko Higashimura",
  "created": {
    "type": "/type/datetime",
    "value": "2019-05-24T01:21:47.045278"
  },
  "last_modified": {
    "type": "/type/datetime",
    "value": "2019-05-24T01:21:47.045278"
  },
  "latest_revision": 1,
  "key": "/authors/OL7527945A",
  "type": {
    "key": "/type/author"
  },
  "revision": 1
}
```
## Endpoint 1C: Volumen de Obras (/authors/{ID}/works.json)

Propósito: Averiguar cuántos libros ha escrito este autor en toda su vida.

Método HTTP: GET

URL Base: https://openlibrary.org/authors/OL3418464A/works.json

Configuración en Thunder Client:
```
{
  "links": {
    "self": "/authors/OL3418464A/works.json?limit=0",
    "author": "/authors/OL3418464A",
    "next": "/authors/OL3418464A/works.json?limit=0&offset=0"
  },
  "size": 1,
  "entries": []
}
```
