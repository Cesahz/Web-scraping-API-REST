```python
import requests
from bs4 import BeautifulSoup
url = "https://books.toscrape.com/"
res = requests.get(url, timeout=10)
res.raise_for_status()
#este es la estructura completa pero en texto simple
html = res.text
#aca es un una estructura interpretada, se puede acceder a sus propiedades como clases,id, etiquetas etc
soup = BeautifulSoup(html,"html.parser")
#buscar etiqueta title, accede solo a su texto, accede al texto sin espacios
print(soup.title)
print(soup.title.text)
print(soup.title.text.strip())
```

```python
import requests
from bs4 import BeautifulSoup

stars = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

url = "https://books.toscrape.com/"
res = requests.get(url, timeout=10)
res.raise_for_status()

soup = BeautifulSoup(res.text, "html.parser")

books = soup.find_all("article", class_="product_pod")

books_info = []

for book in books:
    a = book.find("h3").find("a")
    title = a["title"]
    href = a["href"]
    price_text = book.find("p", class_="price_color").text
    price = float(price_text.replace("£", "").replace("Â","").strip())
    rating_class = book.find("p",class_="star-rating")["class"][1]
    rating = stars.get(rating_class,0)
    fila = (title,href,price,rating)
    books_info.append(fila)
for info in books_info[:4]:
    print(info)

```

```python
def scrapear_libros():
    stars = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
    }
    books_info = []
    for i in range(1,51):
        url = f"https://books.toscrape.com/catalogue/page-{i}.html"
        res = requests.get(url,timeout=10)
        res.raise_for_status()
        soup = BeautifulSoup(res.text,"html.parser")
        books = soup.find_all("article",class_="product_pod")
        
        for book in books:
            a = book.find("h3").find("a")
            title = a["title"]
            href = a["href"]
            price_text = book.find("p", class_="price_color").text
            price = float(price_text.replace("£", "").replace("Â","").strip())
            rating_class = book.find("p",class_="star-rating")["class"][1]
            rating = stars.get(rating_class,0)
            fila = (title,href,price,rating)
            books_info.append(fila)
    return books_info
print(len(scrapear_libros()))

            
                    
```
